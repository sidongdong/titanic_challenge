```python
def piecount(col):
  f,ax=plt.subplots(1,2,figsize=(15,6))  #plt.subplots(열 갯수, 행 갯수, 크기)

  #explode : 1이면 파이차트 분리, 0이면 분리 없음
  #autopct : '%1.1f%%' 앞자리 1자리 이상, 소수점 이하 1자리
  train[col].value_counts().plot.pie(explode=[0.1 for i in range(train[col].nunique())],autopct='%1.1f%%', ax=ax[0],shadow=True) 

  #아래 그래프 둘 중 첫번째 것
  ax[0].set_title(col)
  ax[0].set_ylabel('')

  
  sns.countplot(x=col, data=train, ax=ax[1])
  ax[1].set_title(col)
  plt.show()

piecount('Survived')
```

* 분류 관련 머신 러닝에서는 classifier라는 알고리즘이 필요

* Decision Tree : 분류시 해당 여부에 따라 트리의 특정 분기로 이동 (단계별 결정)

* Random Forest : 랜덤으로 생성된 많은 의사결정 트리 (독립적) -> 개별이 틀려도 평균하여 올바른 결과를 얻을 수 있다

* Extra Tree : 부트스트랩 안함 + 최선의 노드가 아닌 임의의 노드 선택 (extremrly randomized trees)

* kneighbor : k개의 주변 값들을 이용하여 주어진 값을 분류

* 로지스틱 회귀 : 종속변수가 이항분포 (0 또는 1)

* weak classifier : 신뢰도가 떨어지나 합하여 신뢰도 증가 -> strong classifier

* emsemble : 여러 개의 learner를 합하여 더 좋은 결과를 내는 기법으로, 대표적으로 boosting과 bagging이 있음

* boosting : 서로 연결되어 있어 잘 못 분류된 것에 집중하여 중요도를 준다. gradient boosting의 경우, loss function을 최적화하여 weak learner가 만들어지고, 이것이 합쳐져 strong learner을 만듦

* bagging : train data에서 랜덤으로 sub data를 만든 후, 각 sub data에서 독립적으로 만든 모델의 결과를 합쳐 평균값을 낸다.

* XGB : gradient boosting의 발전 버전. 효율을 높이고 paralleization을 통해 연결을 없애 속도 향상. 대량 데이터 처리 가능, 캐시 최적화로 리소스 낭비 줄임

* SVC(support vector classifier) : 분류에 마진을 주어 오차 허용, 교차 검증을 통해 최적 마진 선정

* AdaBoost(adaptive boost) : 부스팅의 일종으로, weak learner 어러 개로 strong learner 생성. 여러 classifier를 결합하여 정확도를 높임. 반복 앙상블. 훈련된 classifier에 가중치를 할당하여 향상시킨다



```python
ran = RandomForestClassifier(random_state = 1)
knn = KNeighborsClassifier()
log = LogisticRegression()
xgb = XGBClassifier()
gbc = GradientBoostingClassifier()
svc = SVC(probability = True)
ext = ExtraTreesClassifier()
ada = AdaBoostClassifier()
gnb = GaussianNB()
gpc = GaussianProcessClassifier()
bag = BaggingClassifier()


# 리스트 준비
models = [ran, knn, log, xgb, gbc, svc, ext, ada, gnb, gpc, bag]         
model_names = ['Random Forest', 'K Nearest Neighbour', 'Logistic Regression', 'XGBoost', 'Gradient Boosting', 'SVC', 'Extra Trees', 'AdaBoost', 'Gaussian Naive Bayes', 'Gaussian Process', 'Bagging Classifier']
scores = {}


#모델을 학습 시키고 교차 검증(트레인 세트를 트레인 세트와 검증 세트로 나누고 훈련된 모델이 검증 세트에 대해 작동하는지 확인) 
for ind, mod in enumerate(models):
    mod.fit(X_train, y_train)
    acc = cross_val_score(mod, X_train, y_train, scoring = "accuracy", cv = 10) #cross validation 10번 시행
    scores[model_names[ind]] = acc
```

#Last Voting

* hard voting : 작은 subset의 1 또는 0에 투표
* soft voting : 각 subset의 확률로 계산

```python
grid_hard = VotingClassifier(estimators = [('Random Forest', ran), 
                                           ('Logistic Regression', log),
                                           ('XGBoost', xgb),
                                           ('Gradient Boosting', gbc),
                                           ('Extra Trees', ext),
                                           ('AdaBoost', ada),
                                           ('Gaussian Process', gpc),
                                           ('SVC', svc),
                                           ('K Nearest Neighbour', knn),
                                           ('Bagging Classifier', bag)], voting = 'hard')

grid_hard_cv = model_selection.cross_validate(grid_hard, X_train, y_train, cv=10)
grid_hard.fit(X_train, y_train)

print("Hard voting on test set score mean: {:.2f}". format(grid_hard_cv['test_score'].mean() * 100))
```

```python
#soft voting
grid_soft = VotingClassifier(estimators = [('Random Forest', ran), 
                                           ('Logistic Regression', log),
                                           ('XGBoost', xgb),
                                           ('Gradient Boosting', gbc),
                                           ('Extra Trees', ext),
                                           ('AdaBoost', ada),
                                           ('Gaussian Process', gpc),
                                           ('SVC', svc),
                                           ('K Nearest Neighbour', knn),
                                           ('Bagging Classifier', bag)], voting = 'soft')

grid_soft_cv = model_selection.cross_validate(grid_soft, X_train, y_train, cv=10)
grid_soft.fit(X_train, y_train)

print("Soft voting on test set score mean: {:.2f}". format(grid_soft_cv['test_score'].mean() * 100))
```
